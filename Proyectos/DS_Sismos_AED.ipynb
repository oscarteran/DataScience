### Análisis de datos exploratorios

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('default')
```

```python
datos = pd.read_csv('database.csv')
datos.head()
```

```python
datos.describe()
```

```python
plt.figure(figsize=(8,5))
plt.plot(datos['Latitude'], datos['Longitude'], '.')
plt.title('Distribución Geoespacial de sismos', fontsize = 16)
plt.xlabel('Latitud [°]')
plt.ylabel('Longitud [°]')
#plt.grid()
plt.show()
```

```python
plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
plt.hist(datos['Magnitude'], bins = 50)
plt.title('Histograma de magnitudes')
plt.xlabel('Magnitud')
plt.ylabel('Frecuencia [n]')

plt.subplot(1,2,2)
plt.hist(datos['Depth'], bins = 50)
plt.title('Histograma de profundidades')
plt.xlabel('Profundidad [Km]')
plt.ylabel('Frecuencia [n]')

plt.show()
```
