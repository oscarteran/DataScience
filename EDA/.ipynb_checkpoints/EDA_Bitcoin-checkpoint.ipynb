```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
sns.set_style('darkgrid')
```

```python
# Opening files
file_path = '../DataBase/BTCUSD_1hr.csv'
df = pd.read_csv(file_path)
df
```

```python
print('Information about the dataset BTCUSD_1hr')
df.info()
```

```python
print('The dimension of the dataset is: \n',
     'Number of rows:    ', df.shape[0],
     '\n Number of columns: ', df.shape[1])
```

```python
print(df.isna().sum())
print('\n The dataset does not contain null vallues')
```

```python
print('Descriptive statistics of the dataframe \n')
df.describe()
```

```python
# We order and craete a new dataframe to plot
df_plot = df.drop(columns=['Unix Timestamp', 'Date', 'Symbol', 'Volume'], axis=1)
df_plot
```

```python
# Begin of data visualization
plt.figure(figsize=(10,7))
date = pd.date_range(start=df['Date'][len(df)-1], periods=len(df), freq='H')
plt.title('Price of bitcoin', fontsize=18)
plt.ylabel('Price [$]')
plt.xlabel('Date')
plt.grid(b=True, color='k', linestyle='--', alpha=0.25)
sns.lineplot(x=date, y=df_plot['Open'], label='Bitcoin')
plt.plot(date[df_plot['Open'].idxmax()], df_plot['Open'].max(), 'k*', markersize=10, label='Maximun value')
plt.plot(date[df_plot['Open'].idxmin()], df_plot['Open'].min(), 'ko', markersize=10, label='Minimun value')
plt.legend()
plt.show()
```
