```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Iris.csv')
df.drop(columns='Id', inplace=True)
df
```

```python
# Ejercicio 1
sns.countplot(x = 'Species', data=df)
plt.title('Frequency of the three species in the Iris Dataset')
plt.ylabel('Frequency')
plt.ylim(0,80)
plt.show()
```

```python
# Ejercicio 2
sns.scatterplot(x='SepalLengthCm', y='SepalWidthCm', data=df, hue='Species', alpha=0.5)
plt.title('Relationship between Sepal Length and Sepal Width', fontsize = 12)
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.show()
```

```python
sns.scatterplot(x='PetalLengthCm', y='PetalWidthCm', data=df, hue='Species', alpha=0.5)
plt.title('Relationship between Sepal Length and Sepal Width', fontsize = 12)
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.show()
```

```python
# Ejercicio 3
for column in df.columns[:-1]:
    sns.histplot(df[column])
    plt.show()
```

```python
# Ejercicio 4
for column in df.columns[:-1]:
    sns.boxplot(data=df[column])
    plt.title(column)
    plt.show()
```

```python
plt.figure(figsize = (8, 5))
sns.boxplot(data = df)
plt.title('Box plot')
plt.show()
```

```python
# Ejercicio 5
corr = df.corr()  #Compute pairwise correlation of columns
sns.heatmap(corr, cmap = 'Blues', annot = True)
```

```python
sns.pairplot(data = df)
plt.show()
```
