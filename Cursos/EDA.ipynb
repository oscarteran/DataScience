```python
df = df.drop(columns=['Domain Code', 'Domain', 'Area Code', 
                      'Element Code', 'Item Code', 'Year Code', 'Flag', 'Flag Description'], axis=1)
df
```

```python
def Count(country):
    if country == 'Argentina':
        return 'AR'
    elif country == 'United States of America':
        return 'USA'
    else:
        return 'BR'
    
df['Area'] = df['Area'].apply(Count)
```
