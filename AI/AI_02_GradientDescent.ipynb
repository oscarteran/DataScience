Gradient descent algorithm.

Optimization of non-convex functions.

```python
# Important libraries
import numpy as np
import scipy as sc
import matplotlib.pyplot as plt
```

```python
# Function to optimize
func = lambda th: np.sin(0.5 * th[0]**2 - 0.25 * th[1]**2 + 3) * np.cos(2*th[0] + 1 - np.e**th[1])

# Number of samples
n = 100

# Vectors
X = np.linspace(-2,2,n)
Y = np.linspace(-2,2,n)

# Grid for 3D Graphic
_X, _Y = np.meshgrid(X, Y) 

creation and calculation of z
Z = np.zeros((n,n))

for ix, x in enumerate(X):
    for iy, y in enumerate(Y):
        Z[iy, ix] = func([x,y])
```

```python
# 2D Graphic 
plt.figure(figsize=(10,7))
plt.contourf(X,Y,Z, levels=100)
plt.colorbar()


np.random.seed(6)
Theta = np.random.rand(2) * 4 - 2
Theta = [-1.9,-1]
_T = np.copy(Theta)

plt.plot(Theta[0], Theta[1], 'o', color='black', label = 'Begin')

# Algorith, parameters
h = 0.01
lear_grade = 0.01
grad = np.zeros(2)

# Gradient descent
for _ in range(1000):
    for it, th in enumerate(Theta):
        _T = np.copy(Theta)
        _T[it] = _T[it] + h
        deriv = func(_T) - func(Theta) / h
        grad[it] = deriv
        
    Theta = Theta - lear_grade*grad  
    
    if (_%1 == 0):
        plt.plot(Theta[0], Theta[1], 'o', color='blue')
        
# 2D Graphic 
plt.plot(Theta[0], Theta[1], 'o', color='red',label='End')
plt.legend()
plt.show()
```

```python
# Final graphic with the calculated points

# If you want to manipulate the graph, uncomment the following line
#%matplotlib auto
fig = plt.figure(figsize=(8,8))
ax = plt.axes(projection='3d')
ax.plot(-1.9,-1,func([-1.9,-1]),'o', color='green', label='Begin', markersize=10)
ax.plot(Theta[0], Theta[1], 'o', color='blue',label='End', markersize=10)
ax.contour3D(_X, _Y, Z, levels=100, alpha=0.5, cmap='inferno')
ax.set_title('Function to optimize', fontsize=18)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
plt.legend()
plt.show()
```
