```python
# We import the library.
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# We open a csv file.
df = pd.read_csv('../DataBase/Data5.csv')
df.drop(columns='Id', inplace=True)
df
```

```python
# Exercise 1. Bar plot from seaborn. 
sns.countplot(x = 'Species', data=df)
plt.title('Frequency of the three species in the Iris Dataset')
plt.ylabel('Frequency')
plt.ylim(0,80)
plt.show()
```

```python
# Exercise 2. Sccater plot with color differentiation.
sns.scatterplot(x='SepalLengthCm', y='SepalWidthCm', data=df, hue='Species', alpha=0.5)
plt.title('Relationship between Sepal Length and Sepal Width', fontsize = 12)
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.show()
```

```python
sns.scatterplot(x='PetalLengthCm', y='PetalWidthCm', data=df, hue='Species', alpha=0.5)
plt.title('Relationship between Sepal Length and Sepal Width', fontsize = 12)
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.show()
```

```python
# Exercise 3.Bar charts.
for column in df.columns[:-1]:
    sns.histplot(df[column])
    plt.show()
```

```python
# Exercise 4. Box plot.
for column in df.columns[:-1]:
    sns.boxplot(data=df[column])
    plt.title(column)
    plt.show()
```

```python
plt.figure(figsize = (8, 5))
sns.boxplot(data = df)
plt.title('Box plot')
plt.show()
```

```python
# Exercise 5. Heatmap of correlation.
corr = df.corr()  #Compute pairwise correlation of columns
sns.heatmap(corr, cmap = 'Blues', annot = True)
```

```python
# Pairplot
sns.pairplot(data = df)
plt.show()
```

```python
# Exercise 16.
np.random.seed(20)
n = 2_000  #size
data = {
    'A': np.random.normal(loc = 100, scale = 1, size = n),
    'B': np.random.normal(loc = 100, scale = 2, size = n),
    'C': np.random.normal(loc = 100, scale = 5, size = n),
    'D': np.random.lognormal(mean = 2, sigma = 1.1, size = n),
    'E': np.random.lognormal(mean = 2, sigma = 1.2, size = n),
    'F': np.random.lognormal(mean = 2, sigma = 1.3, size = n),
    'G': np.random.randn(n),
    'H': np.random.binomial(n = 10, p = 0.8, size = n)
}
 
df = pd.DataFrame(data)
df
```

```python
sns.kdeplot(x = 'A', data = df)
sns.kdeplot(x = 'B', data = df)
sns.kdeplot(x = 'C', data = df)
plt.legend(['A', 'B', 'C'])
plt.grid(linestyle='--', alpha=0.5)
plt.show()
```

```python
# Exercise 17.
sns.kdeplot(x = 'D', data = df)
sns.kdeplot(x = 'E', data = df)
sns.kdeplot(x = 'F', data = df)
plt.legend(['D', 'E', 'F'])
plt.grid(linestyle='--', alpha=0.5)
plt.show()
```

```python
# Exercise 18.
sns.jointplot(x='A', y='B', data=df)
plt.show()

sns.jointplot(x='D', y='F', data=df)
plt.show()
```

```python
for col in df.columns:
    sns.histplot(df[col], kde=True)
    plt.show()
```

```python
# Exercise 20.
sns.violinplot(y='G', data=df)
plt.show()
```

```python
sns.violinplot(y='H', data=df)
plt.show()
```
